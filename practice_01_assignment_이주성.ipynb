### Ïù¥Î°† Î≥µÏäµ

| ÎåÄÌëúÍ∞í | ÌäπÏßï | Ïù¥ÏÉÅÏπò ÎØºÍ∞êÎèÑ |
|--------|------|:-------------:|
| ÏÇ∞Ïà†ÌèâÍ∑† | Î™®Îì† Í∞í Î∞òÏòÅ | Îß§Ïö∞ ÎØºÍ∞ê |
| Ï§ëÏïôÍ∞í | Ï†ïÎ†¨ ÌõÑ Í∞ÄÏö¥Îç∞ Í∞í | Í∞ïÍ±¥Ìï® |
| ÏµúÎπàÍ∞í | Í∞ÄÏû• ÎπàÎ≤àÌïú Í∞í | - |
| Ï†àÏÇ¨ÌèâÍ∑† | ÏñëÎÅù Ï†úÍ±∞ ÌõÑ ÌèâÍ∑† | Ï§ëÍ∞Ñ |
| Í∞ÄÏ§ëÌèâÍ∑† | Í∞ÄÏ§ëÏπò Î∞òÏòÅ ÌèâÍ∑† | Îß§Ïö∞ ÎØºÍ∞ê |

### ÏàòÌñâÌï† ÏûëÏóÖ
1. EMISSIONSÏùò ÏÇ∞Ïà†ÌèâÍ∑†, Ï§ëÏïôÍ∞í, ÏµúÎπàÍ∞í Í≥ÑÏÇ∞
2. 10%, 20% Ï†àÏÇ¨ÌèâÍ∑† Í≥ÑÏÇ∞ Î∞è ÏÇ∞Ïà†ÌèâÍ∑†Í≥º ÎπÑÍµê
3. Ï∞®ÎüâÎì±Í∏â(COMPACT, MID-SIZE, SUV)Ïù∏ Ï∞®ÎüâÏóê ÎåÄÌïú Í∞ÄÏ§ëÌèâÍ∑† Í≥ÑÏÇ∞ (COMPACT 50%, MID-SIZE 30%, SUV 20%)
4. Ï∞®ÎüâÎì±Í∏â(COMPACT, MID-SIZE, SUV)Î≥Ñ ÌèâÍ∑†/Ï§ëÏïôÍ∞í ÎπÑÍµê
5. Í≤∞Í≥º Ìï¥ÏÑù: Ïñ¥Îñ§ ÏÉÅÌô©ÏóêÏÑú Ïñ¥Îñ§ ÎåÄÌëúÍ∞íÏù¥ Ï†ÅÏ†àÌïúÏßÄ Ï†ïÎ¶¨

```python
# TODO 1: Í∏∞Î≥∏ ÎåÄÌëúÍ∞í Í≥ÑÏÇ∞ (ÏÇ∞Ïà†ÌèâÍ∑†, Ï§ëÏïôÍ∞í, ÏµúÎπàÍ∞í)
co2_emission = df['EMISSIONS']
print(f'ÏÇ∞Ïà†ÌèâÍ∑†:{co2_emission.mean():.2f}')
print(f'Ï§ëÏïôÍ∞í:{co2_emission.median():.2f}')
print(f'ÏµúÎπàÍ∞í:{co2_emission.mode().values}') 


# TODO 2: Ï†àÏÇ¨ÌèâÍ∑† Í≥ÑÏÇ∞
print(f'Ï†àÏÇ¨ÌèâÍ∑† (10%):{stats.trim_mean(co2_emission, 0.1):.2f}') # ÏÉÅ,Ìïò 10% Ï†úÍ±∞ ÌõÑ ÌèâÍ∑†


# TODO 3: Ï∞®ÎüâÎì±Í∏â(COMPACT, MID-SIZE, SUV)Ïù∏ Ï∞®ÎüâÏóê ÎåÄÌïú Í∞ÄÏ§ëÌèâÍ∑† Í≥ÑÏÇ∞ (COMPACT 50%, MID-SIZE 30%, SUV 20%)
species_mean = (
    df
    .groupby('VEHICLE CLASS')['EMISSIONS']
    .mean()
    .loc[['COMPACT','MID-SIZE','SUV']]
)
weights = np.array([0.5, 0.3, 0.2])
weighted_mean = np.average(species_mean, weights=weights)
print(f'Í∞ÄÏ§ëÌèâÍ∑†: {weighted_mean:.2f}')


# TODO 4: Ï∞®ÎüâÎì±Í∏â(COMPACT, MID-SIZE, SUV)Î≥Ñ ÌèâÍ∑†/Ï§ëÏïôÍ∞í ÎπÑÍµê
compact = df.loc[df['VEHICLE CLASS'] == 'COMPACT',  'EMISSIONS']
# compact = df[df['VEHICLE CLASS'] == 'COMPACT']['EMISSIONS'] -> locÍ∞Ä Îçî ÏïàÏ†Ñ
mid_size = df.loc[df['VEHICLE CLASS'] == 'MID-SIZE', 'EMISSIONS']
suv      = df.loc[df['VEHICLE CLASS'] == 'SUV',      'EMISSIONS']

comparison = pd.DataFrame({
    'ÌÜµÍ≥ÑÎüâ': ['ÏÇ∞Ïà†ÌèâÍ∑†', 'Ï†àÏÇ¨ÌèâÍ∑†(10%)', 'Ï§ëÏïôÍ∞í'],
    'COMPACT':  [compact.mean(),  stats.trim_mean(compact, 0.1),  compact.median()],
    'MID-SIZE': [mid_size.mean(), stats.trim_mean(mid_size, 0.1), mid_size.median()],
    'SUV':      [suv.mean(),      stats.trim_mean(suv, 0.1),      suv.median()]
})

display(comparison.round(2))


# TODO 5: Í≤∞Í≥º Ìï¥ÏÑù (printÎ¨∏ÏúºÎ°ú ÏûëÏÑ±)
print("COMPACT Îì±Í∏âÏùÄ ÌèâÍ∑†Ïù¥ Ï§ëÏïôÍ∞íÎ≥¥Îã§ ÏïΩÍ∞Ñ ÎÜíÎã§")
print("MID-SIZE Îì±Í∏âÏùÄ ÌèâÍ∑†Ïù¥ Ï§ëÏïôÍ∞íÎ≥¥Îã§ ÎÜíÎã§")
print("SUV Îì±Í∏âÏùÄ ÌèâÍ∑†Ïù¥ ÌèâÍ∑†Ïù¥ Ï§ëÏïôÍ∞íÎ≥¥Îã§ ÎÜíÎã§")
print("ÎåÄÌëúÍ∞í Î∂ÑÏÑùÏóêÏÑúÎäî Ï§ëÏïôÍ∞í ÎòêÎäî Ï†àÏÇ¨ÌèâÍ∑†Ïù¥ Î≥¥Îã§ ÏïàÏ†ïÏ†ÅÏù∏ ÏßÄÌëúÍ∞Ä Îê† Í∞ÄÎä•ÏÑ±Ïù¥ ÏûàÎã§.")
```

## üìù Ïã§Ïäµ Î¨∏Ï†ú 2: ÏÇ∞Ìè¨ÎèÑ Ï∏°ÎèÑ

### Î∞∞Í≤Ω
**COMB (L/100 km)** ÏôÄ **ENGINE SIZE**Ïùò ÏÇ∞Ìè¨Î•º Î∂ÑÏÑùÌï©ÎãàÎã§.

### Ïù¥Î°† Î≥µÏäµ

| ÏßÄÌëú | Í≥ÑÏÇ∞ Î∞©Ïãù | Ïù¥ÏÉÅÏπò ÎØºÍ∞êÎèÑ |
|------|----------|:------------:|
| Î∂ÑÏÇ∞/ÌëúÏ§ÄÌé∏Ï∞® | ÌèâÍ∑† Í∏∞Ï§Ä Ìé∏Ï∞® Ï†úÍ≥± | ÎØºÍ∞ê |
| Î≤îÏúÑ | ÏµúÎåÄ - ÏµúÏÜå | Îß§Ïö∞ ÎØºÍ∞ê |
| IQR | Q3 - Q1 | Í∞ïÍ±¥ |

### Ïù¥ÏÉÅÏπò ÌÉêÏßÄ Î∞©Î≤ï
- **IQR Î∞©Î≤ï**: Q1 - 1.5√óIQR ÎØ∏Îßå ÎòêÎäî Q3 + 1.5√óIQR Ï¥àÍ≥º
- **Z-score Î∞©Î≤ï**: |Z| > 2 ÎòêÎäî 3

### ÏàòÌñâÌï† ÏûëÏóÖ
1. Îëê Î≥ÄÏàòÏùò Î∂ÑÏÇ∞, ÌëúÏ§ÄÌé∏Ï∞®, Î≤îÏúÑ, IQR Í≥ÑÏÇ∞
2. Î≥ÄÎèôÍ≥ÑÏàò(CV) Í≥ÑÏÇ∞ÌïòÏó¨ ÏÉÅÎåÄÏ†Å ÏÇ∞Ìè¨ ÎπÑÍµê
3. IQR Î∞©Î≤ïÍ≥º Z-score Î∞©Î≤ïÏúºÎ°ú Ïù¥ÏÉÅÏπò ÌÉêÏßÄ Î∞è ÎπÑÍµê
4. 2√ó2 ÏãúÍ∞ÅÌôî: Í∞Å Î≥ÄÏàòÏùò ÌûàÏä§ÌÜ†Í∑∏Îû® + Ï∞®ÎüâÎì±Í∏âÎ≥Ñ Î∞ïÏä§ÌîåÎ°Ø
5. Ï∞®ÎüâÎì±Í∏âÎ≥Ñ(COMPACT, MID-SIZE, SUV) Îëê Î≥ÄÏàòÏùò ÌëúÏ§ÄÌé∏Ï∞® ÎπÑÍµê
6. Í≤∞Í≥º Ìï¥ÏÑù

```python
# TODO 1: Îëê Î≥ÄÏàòÏùò ÏÇ∞Ìè¨ Ï∏°ÎèÑ Í≥ÑÏÇ∞
mileage = ['COMB (L/100 km)', 'ENGINE SIZE']

dispersion = pd.DataFrame({
    'Î∂ÑÏÇ∞': df[mileage].var(),
    'ÌëúÏ§ÄÌé∏Ï∞®': df[mileage].std(),
    'Î≤îÏúÑ': df[mileage].max() - df[mileage].min(),
    'IQR': df[mileage].quantile(0.75) - df[mileage].quantile(0.25)
})

display(dispersion.round(3))

# TODO 2: Î≥ÄÎèôÍ≥ÑÏàò(CV) Í≥ÑÏÇ∞
print("Cv:")
cv = df[mileage].std() / df[mileage].mean() * 100
display(cv.round(3))


# TODO 3: Ïù¥ÏÉÅÏπò ÌÉêÏßÄ (IQR, Z-score)
Q1 = df['COMB (L/100 km)'].quantile(0.25) 
Q3 = df['COMB (L/100 km)'].quantile(0.75)
IQR = Q3 - Q1
iqr_lower = Q1 - 1.5 * IQR
iqr_upper = Q3 + 1.5 * IQR

iqr_outliers = df['COMB (L/100 km)'][(df['COMB (L/100 km)'] < iqr_lower) | (df['COMB (L/100 km)'] > iqr_upper)] 
print('COMB (L/100 km)Ïùò IQR Ïù¥ÏÉÅÏπò:') 
print('COMB (L/100 km)ÏùòIQR Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(iqr_outliers))

## Z-Score Î∞©Î≤ï 
z_scores = (df['COMB (L/100 km)'] - df['COMB (L/100 km)'].mean()) / df['COMB (L/100 km)'].std() 
z_outliers_2 = df['COMB (L/100 km)'][np.abs(z_scores) > 2] 
z_outliers_3 = df['COMB (L/100 km)'][np.abs(z_scores) > 3] 
print('COMB (L/100 km)Ïùò Z-score Ïù¥ÏÉÅÏπò:') 
print('COMB (L/100 km)Ïùò |Z-score| > 2 Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(z_outliers_2)) 
print('COMB (L/100 km)Ïùò |Z-score| > 3 Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(z_outliers_3))


# ENGINE SIZE 
Q1 = df['ENGINE SIZE'].quantile(0.25) 
Q3 = df['ENGINE SIZE'].quantile(0.75) 
IQR = Q3 - Q1 
iqr_lower = Q1 - 1.5 * IQR 
iqr_upper = Q3 + 1.5 * IQR 
iqr_outliers = df['ENGINE SIZE'][(df['ENGINE SIZE'] < iqr_lower) | (df['ENGINE SIZE'] > iqr_upper)] 
print('ENGINE SIZEÏùò IQR Ïù¥ÏÉÅÏπò:') 
print('ENGINE SIZEÏùò IQR Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(iqr_outliers))

## Z-Score Î∞©Î≤ï 
z_scores = (df['ENGINE SIZE'] - df['ENGINE SIZE'].mean()) / df['ENGINE SIZE'].std() 
z_outliers_2 = df['ENGINE SIZE'][np.abs(z_scores) > 2] 
z_outliers_3 = df['ENGINE SIZE'][np.abs(z_scores) > 3] 
print('ENGINE SIZEÏùò Z-score Ïù¥ÏÉÅÏπò:') 
print('ENGINE SIZEÏùò |Z-score| > 2 Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(z_outliers_2)) 
print('ENGINE SIZEÏùò |Z-score| > 3 Ïù¥ÏÉÅÏπò Í∞úÏàò:', len(z_outliers_3))


# TODO 4: 2√ó2 ÏãúÍ∞ÅÌôî: Í∞Å Î≥ÄÏàòÏùò ÌûàÏä§ÌÜ†Í∑∏Îû® + Ï∞®ÎüâÎì±Í∏âÎ≥Ñ Î∞ïÏä§ÌîåÎ°Ø
classes = ['COMPACT', 'MID-SIZE', 'SUV']

df2 = df[df['VEHICLE CLASS'].isin(['COMPACT','MID-SIZE','SUV'])] 

fig, ax = plt.subplots(2, 2, figsize=(12, 8)) 
# (0,0) COMB ÌûàÏä§ÌÜ†Í∑∏Îû®
ax[0,0].hist(df['COMB (L/100 km)'].dropna(), bins=30) 
ax[0,0].set_title('COMB (L/100 km) Histogram') 
ax[0,0].set_xlabel('COMB (L/100 km)') 
ax[0,0].set_ylabel('Count') 

# (0,1) ENGINE SIZE ÌûàÏä§ÌÜ†Í∑∏Îû® 
ax[0,1].hist(df['ENGINE SIZE'].dropna(), bins=30) 
ax[0,1].set_title('ENGINE SIZE Histogram') 
ax[0,1].set_xlabel('ENGINE SIZE') 
ax[0,1].set_ylabel('Count') 

# (1,0) Ï∞®ÎüâÎì±Í∏âÎ≥Ñ COMB Î∞ïÏä§ÌîåÎ°Ø 
df[df['VEHICLE CLASS'].isin(classes)].boxplot(
    column='COMB (L/100 km)',
    by='VEHICLE CLASS',
    ax=ax[1,0]
)
ax[1,0].set_title('COMB (L/100 km) by VEHICLE CLASS') 
ax[1,0].set_xlabel('VEHICLE CLASS') 
ax[1,0].set_ylabel('COMB (L/100 km)') 

# (1,1) Ï∞®ÎüâÎì±Í∏âÎ≥Ñ ENGINE SIZE Î∞ïÏä§ÌîåÎ°Ø 
df[df['VEHICLE CLASS'].isin(classes)].boxplot(
    column='ENGINE SIZE',
    by='VEHICLE CLASS',
    ax=ax[1,1]
)

ax[1,1].set_title('ENGINE SIZE by VEHICLE CLASS') 
ax[1,1].set_xlabel('VEHICLE CLASS') 
ax[1,1].set_ylabel('ENGINE SIZE') 


plt.tight_layout() 
plt.show()


# TODO 5: Ï∞®ÎüâÎì±Í∏âÎ≥Ñ(COMPACT, MID-SIZE, SUV) ÌëúÏ§ÄÌé∏Ï∞® ÎπÑÍµê
std_class = (
    df[df['VEHICLE CLASS'].isin(['COMPACT','MID-SIZE','SUV'])]
    .groupby('VEHICLE CLASS')[mileage]
    .std()
)
display(std_class.round(3))

# TODO 6: Í≤∞Í≥º Ìï¥ÏÑù
print("Ï∞®Îüâ Îì±Í∏âÏù¥ Ï¶ùÍ∞ÄÌï†ÏàòÎ°ù Ïó∞ÎπÑÏùò Î≥ÄÎèôÏÑ±ÏùÄ Ï¶ùÍ∞ÄÌïúÎã§.")
print("ÌäπÌûà SUVÏóêÏÑú Ïó∞Î£å ÏÜåÎπÑÎüâÏùò Ìé∏Ï∞®Í∞Ä Í∞ÄÏû• ÌÅ¨Í≤å ÎÇòÌÉÄÎÇ¨Îã§.")
```

## üìù Ïã§Ïäµ Î¨∏Ï†ú 3: Î∂ÑÌè¨Ïùò ÌòïÌÉú (ÏôúÎèÑ, Ï≤®ÎèÑ)

### Î∞∞Í≤Ω
4Í∞ÄÏßÄ ÏàòÏπòÌòï Î≥ÄÏàòÏùò **ÏôúÎèÑÏôÄ Ï≤®ÎèÑ**Î•º Î∂ÑÏÑùÌïòÏó¨ Î∂ÑÌè¨ ÌòïÌÉúÎ•º ÌååÏïÖÌï©ÎãàÎã§.

### Ïù¥Î°† Î≥µÏäµ

**ÏôúÎèÑ**
| ÏôúÎèÑ | ÏùòÎØ∏ | ÌèâÍ∑† vs Ï§ëÏïôÍ∞í |
|------|------|---------------|
| = 0 | ÎåÄÏπ≠ | ÌèâÍ∑† ‚âà Ï§ëÏïôÍ∞í |
| > 0 | Ïò§Î•∏Ï™Ω Íº¨Î¶¨ | ÌèâÍ∑† > Ï§ëÏïôÍ∞í |
| < 0 | ÏôºÏ™Ω Íº¨Î¶¨ | ÌèâÍ∑† < Ï§ëÏïôÍ∞í |


**Ï¥àÍ≥º Ï≤®ÎèÑ**
| Ï≤®ÎèÑ | ÏùòÎØ∏ |
|------|------|
| = 0 | Ï†ïÍ∑úÎ∂ÑÌè¨ÏôÄ ÎèôÏùº |
| > 0 | Íº¨Î¶¨ ÎëêÍ∫ºÏõÄ (Í∑πÎã®Í∞í ÎßéÏùå) |
| < 0 | Íº¨Î¶¨ ÏñáÏùå (Í∑πÎã®Í∞í Ï†ÅÏùå) |

### ÏàòÌñâÌï† ÏûëÏóÖ
1. 4Í∞ú Î≥ÄÏàòÏùò ÏôúÎèÑ/Ï≤®ÎèÑ Í≥ÑÏÇ∞: ENGINE SIZE, COMB (L/100 km), COMB (mpg), EMISSIONS
2. Í≤∞Í≥ºÎ•º DataFrameÏúºÎ°ú Ï†ïÎ¶¨ (ÏôúÎèÑ/Ï≤®ÎèÑ Ìï¥ÏÑù Ïª¨Îüº Ìè¨Ìï®)
3. 2√ó2 ÌûàÏä§ÌÜ†Í∑∏Îû®+KDE ÏãúÍ∞ÅÌôî (Ï†úÎ™©Ïóê ÏôúÎèÑ/Ï≤®ÎèÑ ÌëúÏãú)
    - Í∞Å Í∑∏ÎûòÌîÑÏóê ÌèâÍ∑†(Îπ®Í∞ï)Í≥º Ï§ëÏïôÍ∞í(Ï¥àÎ°ù) ÏàòÏßÅÏÑ† ÌëúÏãú
